```python
import pandas as pd

# Load data from both schools_complete.csv and students_complete.csv
schools_file_path = r'C:\Users\n.rennie\Documents\pandas-challenge\Starter_Code\PyCitySchools\Resources\schools_complete.csv'
students_file_path = r'C:\Users\n.rennie\Documents\pandas-challenge\Starter_Code\PyCitySchools\Resources\students_complete.csv'
schools_data = pd.read_csv(schools_file_path)
students_data = pd.read_csv(students_file_path)

# Calculate budget per student for each school
schools_data['Per Student Budget'] = schools_data['budget'] / schools_data['size']

# Define spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create "Spending Range" column using pd.cut
schools_data['Spending Range'] = pd.cut(schools_data['Per Student Budget'], bins=spending_bins, labels=labels)

# Group students data by school name
grouped_students = students_data.groupby('school_name')

# Calculate average scores and passing percentages by spending range
spending_performance = pd.DataFrame({
    'Average Math Score': grouped_students['math_score'].mean(),
    'Average Reading Score': grouped_students['reading_score'].mean(),
    '% Passing Math': (students_data[students_data['math_score'] >= 70].groupby('school_name')['Student ID'].count() / grouped_students['Student ID'].count()) * 100,
    '% Passing Reading': (students_data[students_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count() / grouped_students['Student ID'].count()) * 100,
    '% Overall Passing': (students_data[(students_data['math_score'] >= 70) & (students_data['reading_score'] >= 70)].groupby('school_name')['Student ID'].count() / grouped_students['Student ID'].count()) * 100
})

# Merge school data with spending performance data
spending_performance = pd.merge(schools_data, spending_performance, on='school_name')

# Group by spending range and calculate average scores and passing percentages
spending_performance_summary = spending_performance.groupby('Spending Range', observed=False).agg({
    'Average Math Score': 'mean',
    'Average Reading Score': 'mean',
    '% Passing Math': 'mean',
    '% Passing Reading': 'mean',
    '% Overall Passing': 'mean'
})

# Display the table
spending_performance_summary
```

```python
# Calculate mean scores per spending range
spending_math_scores = spending_performance_summary["Average Math Score"]
spending_reading_scores = spending_performance_summary["Average Reading Score"]
spending_passing_math = spending_performance_summary["% Passing Math"]
spending_passing_reading = spending_performance_summary["% Passing Reading"]
overall_passing_spending = spending_performance_summary["% Overall Passing"]

# Display the results
print("Mean Math Scores per Spending Range:")
print(spending_math_scores)
print("\nMean Reading Scores per Spending Range:")
print(spending_reading_scores)
print("\nMean Passing Math Percentage per Spending Range:")
print(spending_passing_math)
print("\nMean Passing Reading Percentage per Spending Range:")
print(spending_passing_reading)
print("\nMean Overall Passing Percentage per Spending Range:")
print(overall_passing_spending)
```

```python
# Create spending summary DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display the spending summary DataFrame
print(spending_summary)
```
